## WordClouds

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
alice_novel = open('alice_novel.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
alice_wc = WordCloud(background_color='white',
                    max_words=2000,
                    stopwords=stopwords)
alice_wc.generate(alice_novel)
```

```python
fig = plt.figure(figsize=(9,10))
plt.imshow(alice_wc, interpolation='bilinear') #interpolation buat smoothing chart
plt.axis('off')
plt.show()
```

```python
stopwords.add('said')
```

## Regression Plot

bedanya korelasi dengan regresi:
- korelasi data bergerak bersama, regresi ada sebab akibat.
- korelasi x dan y bisa ditukar, regresi x dan y tidak bisa ditukar
- korelasi datanya direpresentasikan oleh titik, kalau regresi pakai garis

```python
import seaborn as sns
```

```python
plt.figure(figsize=(7,5))
sns.regplot(x='year', y='total', data=df_total, color='pink', marker='+')
plt.title('Regression Plot Total Immigrants 1980-2013')
```

```python
df_incidents = pd.read_csv('Police_Department_Incidents_-_Previous_Year__2016_.csv')
```

```python
df_incidents
```

```python
# mapping 100 row pertama dari tabel insiden
df_incidents_100 = df_incidents.iloc[0:100, :]
df_incidents_100
```

```python
#San Fransisco
latitude = 37.77
longitude = -122.42
```

```python
sf_map = folium.Map(location=[latitude, longitude], zoom_start=12)
sf_map
```

```python
sf_map = folium.Map(location=[latitude, longitude], zoom_start=12)
sf_map
incidents = folium.map.FeatureGroup()
#x longitude/bujur, y latitude/lintang (saat dimapping)
for lat,lon in zip(df_incidents_100['Y'], df_incidents_100['X']):
    incidents.add_child(
        folium.CircleMarker(
            [lat,lon],
            radius=5,
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.6
        )
    )

sf_map.add_child(incidents)
```

```python
sf_map = folium.Map(location=[latitude, longitude], zoom_start=12)
sf_map
incidents = folium.map.FeatureGroup()
#x longitude/bujur, y latitude/lintang (saat dimapping)
for lat,lon,label in zip(df_incidents_100['Y'], df_incidents_100['X'], df_incidents_100.Category):
    incidents.add_child(
        folium.CircleMarker(
            [lat,lon],
            radius=5,
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.6,
            popup=label
        )
    )

sf_map.add_child(incidents)
```

```python
#clustering semua data sesuai kemiripan
from folium import plugins
sf_map = folium.Map(location=[latitude, longitude], zoom_start=12)

incidents = plugins.MarkerCluster().add_to(sf_map)

for lat,lon,label in zip(df_incidents_100['Y'], df_incidents_100['X'], df_incidents_100.Category):
    incidents.add_child(
        folium.CircleMarker(
            [lat,lon],
            radius=5,
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.6,
            popup=label
        )
    )

sf_map.add_child(incidents)
```
