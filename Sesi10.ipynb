```python
z_score = 1.96
low_ci_fm=pro_fm-(z_score*stderror_fm)
up_ci_fm=pro_fm+(z_score*stderror_fm)
print(low_ci_fm, up_ci_fm)
```

```python
n=30
sample_means=[]
for sample in range(0,250):
    sample_values=df_mass['people_mass'].sample(n)
    sample_mean=sample_values.mean()
    sample_means.append(sample_mean)
```

```python
population_mean
```

```python
std_of_sample_means=np.std(sample_means)
std_of_sample_means
```

### Standard Error

```python
standard_eror=population_std/np.sqrt(n)
standard_eror
```

```python
std_of_sample_means
```

kesimpulan: sample memiliki karakteristik yang serupa dengan populasi

### VISUALISASI DAERAH CONFIDENCE INTERVAL

```python
standard_eror=population_std/np.sqrt(n)
standard_eror
```

```python
stats.norm.cdf(2.17)
```

```python
z = 1.96
```

```python
lower_bound = mean_of_sample_means - z * standard_eror
lower_bound
```

```python
upper_bound = mean_of_sample_means + z *standard_eror
upper_bound
```

```python
kde=stats.gaussian_kde(sample_means)
pos = np.linspace(np.min(sample_means),np.max(sample_means),10000)
plt.plot(pos,kde(pos))
shade=np.linspace(lower_bound,upper_bound,300)
plt.fill_between(shade,kde(shade), alpha=0.5)
plt.xlabel('sample mass(pounds)')
plt.ylabel('frequency')
plt.axvline(x=mean_of_sample_means,color='black')
plt.text(x=152,y=0.01,s='shaded area \npart of the \nconfidence intervals \n95%')
```

### HYPOTHESIS TESTING

+ if p-value > alpha : fail to reject null hypot
+ if p-value <=alpha: reject the null hypot

alpha biasanya 0.05 atau 5% karena confidence level biasanya 95%.

```python
pd.read_csv('heart.csv')
```

```python
df = pd.read_csv('nhanes_2015_2016.csv')
df
```

Apakah BMI laki laki > perempuan

+ h0 = rata2 BMI lakiki tidak lebih tinggi signifikan dari perempuan
+ h1 = rata2 bmi lakilaki lebih tinggi signifikan dari perempuan

h0:meanL<=meanP

h1:meanL>meanP

```python
df.RIAGENDR
```

```python
females = df[df.RIAGENDR==2]
males = df[df.RIAGENDR==1]
```

```python
n_females =len(females)
mean_females=females['BMXBMI'].mean()
std_females=females['BMXBMI'].std()
print(n_females, mean_females, std_females)
```

```python
n_males =len(males)
mean_males=males['BMXBMI'].mean()
std_males=males['BMXBMI'].std()
print(n_males, mean_males, std_males)
```

```python
import statsmodels.api as sm
```

```python
# karena confidence level 95%
alpha = 0.05
```

```python
bmi_hypo=sm.stats.ztest(males['BMXBMI'].dropna(),females['BMXBMI'].dropna(),alternative='larger')
bmi_hypo
```

```python
if bmi_hypo[1]>alpha:
    print('Accept Null Hypothesis (h0) mean of males bmi is not larger than females')
else:
    print('reject null hypothesis(h0)mean of males bmi is larger than females')
```

```python
bmi_hypo_2s=sm.stats.ztest(males['BMXBMI'].dropna(),females['BMXBMI'].dropna(),alternative='two-sided')
bmi_hypo_2s
```

```python
if bmi_hypo_2s[1]>alpha:
    print('Accept Null Hypothesis (h0)')
else:
    print('reject null hypothesis(h0)')
```

### Hypothesis Testing Population Proportion

```python
n = 1018
pro_null = 0.52
pro_alt = 0.56
```

```python
pro_hypo = sm.stats.proportions_ztest(n*pro_alt,n,pro_null,alternative='larger')
pro_hypo
```

```python
if pro_hypo[1]>alpha:
    print('Accept Null Hypothesis (h0)')
else:
    print('reject null hypothesis(h0)')
```
