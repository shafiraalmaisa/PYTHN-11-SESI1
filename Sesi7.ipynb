## Use Case 3 : Compare top 5 Countries that has most imigrations to Canada

```python
df_can.sort_values(by='Total',ascending=False, axis=0, inplace=True)
df_top5=df_can.head()
df_top5 = df_top5[years].transpose()
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(14,8))

plt.title('Imigrants Trend of Top 5 Countries')
plt.ylabel('Number of Imigrants')
plt.xlabel('Years')
plt.show()
```

## Histogram

nampilin distribusi data dan frekuensi untuk data range. bedanya sama bar chart, bar chart pakai data diskrit kalau histogram pakai data kontinu

## Use Case 4 : Distribusi populasi imigran di 2013

```python
df_can['2013'].head()
```

```python
# np.histogram return 2 values
count, bin_edges = np.histogram(df_can['2013'])
print(count) #frequency
print(bin_edges) #range dari tiap bar, defaultnya 10
```

```python
# Plot pakai scripting layer
df_can['2013'].plot(kind='hist')
plt.title('Histogram of Immigration from 195 countries in 2013')
plt.ylabel('Number of Countries(freq)')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
count, bin_edges = np.histogram(df_can['2013'])

df_can['2013'].plot(kind='hist', figsize=(8,5), xticks=bin_edges)
plt.title('Histogram of Immigration from 195 countries in 2013')
plt.ylabel('Number of Countries(freq)')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
df_can.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
# persebaran data imigrasi dari beberapa negara
df_can.loc[['Denmark', 'Norway', 'Sweden'], years].T.plot(kind='hist')
```

```python
df_T = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].T
count, bin_edges = np.histogram(df_T, 15)

#unstacked histogram
df_T.plot(kind='hist', figsize=(10,6), bins=15, alpha=0.6,xticks=bin_edges, color=['cyan', 'magenta','grey'])
plt.show()
```

## PR LAGI NIH

scripting layer untuk distribusi imigram dari greece albania dan bulgaria 1980-2013. 15 bins dan alpha 0.35

```python
#sini codenya
df_PR = df_can.loc[['Greece', 'Albania', 'Bulgaria'], years].T
count, bin_edges = np.histogram(df_PR, 15)

#unstacked histogram
df_PR.plot(kind='hist', figsize=(10,6), bins=15, alpha=0.35, xticks=bin_edges, color=['cyan', 'magenta','grey'])
plt.show()
```

## Bar Charts

```python
df_ice = df_can.loc['Iceland', years]
df_ice.head()
```

```python
df_ice.plot(kind='bar')
plt.title('Iceland immigrants 1980-2013')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_ice.plot(kind='bar',figsize=(10,6), rot=90)
plt.title('Iceland immigrants 1980-2013')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.annotate('',
            xy=(32,70),
            xytext=(28,20),
            xycoords='data',
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='black', lw=2))
plt.annotate('2008-2011 Financial Crisis',
            xy = (28,30),
            rotation = 72.5,
            va ='bottom',
            ha='left')
plt.show()
```
